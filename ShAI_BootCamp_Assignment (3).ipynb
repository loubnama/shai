```python
import pandas as pd

df = pd.read_csv("/Users/lubna/Downloads/Salaries.csv")

colum = ["JobTitle","BasePay"]

group = df[colum].groupby("BasePay")

out = group.agg({
    "BasePay": ["mean", "median", "min", "max", "std", "count"]
})

print(out)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/lubna/Downloads/Salaries.csv")
selected_columns = ['BasePay','OvertimePay']
correlation = df[selected_columns].corr()
print(correlation)
plt.scatter(df["BasePay"], df["OvertimePay"])
plt.title("Scatter Plot for BasePay vs OvertimePay")
plt.xlabel("BasePay")
plt.ylabel("OvertimePay")
plt.show()

```
